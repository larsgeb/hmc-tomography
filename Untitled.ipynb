```python
import hmc_tomography
import matplotlib.pyplot as plt
```

```python
distribution = hmc_tomography.Distributions.Himmelblau(temperature=100)
```

```python
sampler.plot_stepsizes()
```

```python
sampler.plot_acceptance_rate()
```
